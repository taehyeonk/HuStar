```python
plu = raw_plu.copy()
plu['Measurement date'] = pd.to_datetime(plu['Measurement date'])
```

```python
idx_2017 = plu['Measurement date'].apply(lambda x: x.year == 2017)
print(idx_2017)

plu.loc[idx_2017, :]
```

```python
idx_NoFeb = plu['Measurement date'].dt.month > 2# We can use other operator like >, >=, <, <=
idx_2018 = plu['Measurement date'].dt.weekday == 0 # 0 is monday, 6 is sunday
idx = idx_NoFeb & idx_2018
plu.loc[idx,:]
```
